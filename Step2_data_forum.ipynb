```python

#Data FROM Forums
import pandas as pd
import numpy as np
import ast
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

def import_data(file):
    parkinson_data = pd.read_csv(file)
    parkinson_data['objectivity_proba'] = 0
    return parkinson_data
```

```python
import re

REPLACE_NO_SPACE = re.compile("[.;:!\'?,\"()\[\]]")
REPLACE_WITH_SPACE = re.compile("(<br\s*/><br\s*/>)|(\-)|(\/)")

def pre_process(parkinson_data):
    parkinson_data.dropna(inplace=True)
    parkinson_data.drop_duplicates(inplace=True)
    parkinson_data.comments = list(map(lambda value: ast.literal_eval(str(value).replace('\\n',"").replace("\\t","")),parkinson_data.comments))
    parkinson_data.comments = list(map(lambda row : [row.remove(array) if 'Posts' in array else array for array in row], parkinson_data.comments))  
    parkinson_data.comments = list(map(lambda row : [ REPLACE_NO_SPACE.sub("", str(array).lower()) for array in row], parkinson_data.comments))  
    parkinson_data.comments = list(map(lambda row : [ REPLACE_WITH_SPACE.sub(" ", str(array)) for array in row], parkinson_data.comments)) 
    return parkinson_data


def arrange_data_array(data):
    data_array={}
    for i in range(len(data.comments)):
        print(i)
    return data
```
